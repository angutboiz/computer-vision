```python
from image_tool import *
from PIL import Image
```

```python
my_path = os.getcwd() + "/img/img (1).jpg"
img = load_image(my_path)
display(img)
```

```python
region = (100,120,200,205)

cropped_image = img.crop(region)
display(cropped_image)
```

```python
paste_position = (30,60)
img.paste(cropped_image,paste_position)
display(img)
```

```python
rotate_image = img.rotate(90)
display(rotate_image)
```

```python
transpose_img = img.transpose(Image.Transpose.ROTATE_90)
display(transpose_img)
```

```python
gray_img = img.convert("L")
display(gray_img)
```

```python
red_band,green_band,blue_band = img.split()
display(red_band)
display(blue_band)
display(green_band)

```

```python
merged_img = Image.merge("RGB",(green_band,blue_band,red_band))
display(merged_img)
```
