```python
from PIL import Image
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
my_path = os.getcwd() + "/img/img (1).jpg"
im = Image.open(my_path)
```

```python
img_gray = im.convert('L')
plt.imshow(img_gray, cmap='gray')
plt.imshow(img_gray)
```

```python
im_array = np.array(img_gray)
print(im_array)
```

```python
plt.figure()
plt.hist(im_array.flatten(),bins=128)
plt.title("Biểu đồ histogram của ảnh")
plt.xlabel("Giá trị của pixel")
plt.ylabel("Số lượng")
plt.show()
```

```python
#làm nổi bật đường biên
plt.figure()
plt.gray()
plt.contour(img_gray,origin="image")
plt.show()
```

```python
im1 = Image.open(os.getcwd() + "/img/img (4).jpg").convert("L")
plt.imshow(im1)
#tạo 1 hình vẽ mới
plt.figure()

#đặt màu sắc là màu xám
plt.gray()

#vẽ biểu đồ đường biên (contour) của ảnh
plt.contour(im1,origin="image")

#đảm bảo tỷ lệ trục x và trục y trên biểu đồ là bằng nhau
plt.axis("equal")


im1_array = np.array(im1)
plt.figure()
plt.hist(im1_array.flatten(),bins=128)
plt.title("Biểu đồ histogram của ảnh")
plt.xlabel("Giá trị của pixel")
plt.ylabel("Số lượng")
plt.show()
```
