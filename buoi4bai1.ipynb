```python
from PIL import Image
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
my_path = os.getcwd() + "/img/img (1).jpg"
im = Image.open(my_path)
```

```python
img_gray = im.convert('L')
plt.imshow(img_gray, cmap='gray')
plt.imshow(img_gray)
```

```python
im_array = np.array(img_gray)
print(im_array)
```

```python
plt.figure()
plt.hist(im_array.flatten(),bins=128)
plt.title("Biu  histogram ca nh")
plt.xlabel("Gi tr ca pixel")
plt.ylabel("S lng")
plt.show()
```

```python
#lm ni bt ng bin
plt.figure()
plt.gray()
plt.contour(img_gray,origin="image")
plt.show()
```

```python
im1 = Image.open(os.getcwd() + "/img/img (4).jpg").convert("L")
plt.imshow(im1)
#to 1 hnh v mi
plt.figure()

#t mu sc l mu xm
plt.gray()

#v biu  ng bin (contour) ca nh
plt.contour(im1,origin="image")

#m bo t l trc x v trc y trn biu  l bng nhau
plt.axis("equal")


im1_array = np.array(im1)
plt.figure()
plt.hist(im1_array.flatten(),bins=128)
plt.title("Biu  histogram ca nh")
plt.xlabel("Gi tr ca pixel")
plt.ylabel("S lng")
plt.show()
```
